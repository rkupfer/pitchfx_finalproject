```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle
```

```python
df = pd.read_csv("selected_pitches.csv")
```

```python
mask = df.pitch_res == "C"
df = df[mask]

plot_df = df[["px", "pz"]]

heatmapc, xedges, yedges = np.histogram2d(plot_df.px, plot_df.pz, bins=(64,64))
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

#heatmapb, xedges, yedges = np.histogram2d(plot_dfb.px, plot_dfb.pz, bins=(64,64))
#extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]


# Plot heatmap
plt.clf()
someX, someY = 0, 2.5
fig,ax = plt.subplots()
currentAxis = plt.gca()
currentAxis.add_patch(Rectangle((someX - 0.6, someY - 1), 1.2, 2,
                      alpha=1, facecolor='none'))
currentAxis.add_patch(Rectangle((someX - 1, someY - 1), 1.6, 2,
                      alpha=1, facecolor='none'))
plt.title('All Pitches')
plt.ylabel('y')
plt.xlabel('x')
plt.imshow(heatmapc.T, cmap=plt.cm.Reds, extent=extent)
#plt.imshow(heatmapb.T, cmap=plt.cm.Blues, extent=extent)
#plt.hold(True)
plt.show()
```
